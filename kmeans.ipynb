```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('numpy-images\car-photo.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB

# Reshape the image to a 2D array of pixels
pixels = image.reshape((-1, 3))

# Apply K-means clustering
kmeans = KMeans(n_clusters=10)  # You can adjust the number of clusters
kmeans.fit(pixels)

# Get the labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Assign each pixel to its cluster centroid
segmented_image = centroids[labels].reshape(image.shape)

# Display the original and segmented images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(segmented_image.astype(np.uint8))
plt.title('Segmented Image')
plt.axis('off')

plt.show()
```
