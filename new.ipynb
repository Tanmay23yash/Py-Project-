```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Line plot
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
plt.subplot(3, 3, 1)
plt.plot(x, y)
plt.title('Line Plot')

# 2. Scatter plot
data = np.random.rand(30, 2)
plt.subplot(3, 3, 2)
plt.scatter(data[:, 0], data[:, 1])
plt.title('Scatter Plot')

# 3. Bar plot
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 5]
plt.subplot(3, 3, 3)
plt.bar(categories, values)
plt.title('Bar Plot')

# 4. Histogram
random_data = np.random.randn(1000)
plt.subplot(3, 3, 4)
plt.hist(random_data, bins=20, color='green', edgecolor='black')
plt.title('Histogram')

# 5. Pie chart
sizes = [15, 30, 45, 10]
labels = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
plt.subplot(3, 3, 5)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')

# 6. Box plot
data_box = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.subplot(3, 3, 6)
plt.boxplot(data_box, labels=['Box 1', 'Box 2', 'Box 3'])
plt.title('Box Plot')

# 7. 3D plot
ax = plt.subplot(3, 3, 7 , projection = '3d')
x_3d = np.random.rand(20)
y_3d = np.random.rand(20)
z_3d = np.random.rand(20)
ax.scatter(x_3d, y_3d, z_3d)
ax.set_title('3D Scatter Plot')

plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Setting the axes projection as polar
plt.axes(projection='polar')

# Setting the radius
r = 2

# Creating an array containing the radian values
rads = np.arange(0, (2 * np.pi), 0.01)

# Plotting the circle
for rad in rads:
    plt.polar(rad, r, 'g.')

# Display the Polar plot

ax.tick_params(axis='both', colors='red', labelsize=10)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function for the ellipse
def ellipse(theta, a, b, h, k):
    return a * np.cos(theta) + h, b * np.sin(theta) + k

# Generate the y coordinates for the plot
theta = np.linspace(0, 2 * np.pi, 1000)
x = ellipse(theta, 3, 2, 0, 0)[0]
y = ellipse(theta, 3, 2, 0, 0)[1]

# Create the plot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_aspect('equal')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Ellipse')

# Show the plot
plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set(style='whitegrid', palette='muted')

# Generate a random 15 x 15 grid of data
np.random.seed(sum(map(ord, "corr")))
data = np.random.randn(15, 15)

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
f, ax = plt.subplots(figsize=(6, 5))
sns.heatmap(data, cmap=cmap, cbar=False, center=0,
            square =True, linewidths=.5, cbar_kws={"shrink": .5})
ax.set_title("Custom Heatmap");
```
